```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data = pd.read_csv('datasettaylor - Sheet1.csv')
data.head()
```

```python
data.tail()
```

UNIVARIATE ANALYSIS

1. Categorical


```python
data['album'].value_counts().plot(kind="pie",autopct="%1.2f%%")
plt.title('Size of albums')
```

```python
data['album'].value_counts()
```

```python
data['album'].value_counts().plot(kind="bar")
plt.title('No. of songs in each album')
```

2. Numerical

```python
sns.distplot(data['popularity'])
plt.title('Distribution of rating')
```

```python
data['popularity'].skew()
```

```python
sns.distplot(data['acousticness'])
plt.title('Distribution of acousticness')
```

```python
data['acousticness'].skew()
```

```python
data['acousticness'].describe()
```

```python
sns.boxplot(data['loudness'])
plt.title('Box Plot of loudness')
```
